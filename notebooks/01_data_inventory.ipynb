## Análisis de los datos de 2023

```python
# Importar librerías principales
import pandas as pd
```

```python
# Listado de rutas de archivos Excel en la carpeta raw
from pathlib import Path

RAW_PATH = Path("../data/raw")

excel_files = list(RAW_PATH.rglob("*.xlsx")) + list(RAW_PATH.rglob("*.xls"))

len(excel_files), excel_files
```

```python
# Creación de un inventario de los archivos Excel encontrados
inventory = []

for file in excel_files:
    try:
        df = pd.read_excel(file)
        
        inventory.append({
            "file_name": file.name,
            "file_path": str(file),
            "folder": file.parent.name,
            "n_rows": df.shape[0],
            "n_columns": df.shape[1],
            "columns": list(df.columns),
            "n_missing_total": df.isna().sum().sum(),
            "missing_percentage": df.isna().sum().sum() / (df.shape[0] * df.shape[1]) * 100
        })
        
    except Exception as e:
        inventory.append({
            "file_name": file.name,
            "file_path": str(file),
            "folder": file.parent.name,
            "n_rows": None,
            "n_columns": None,
            "columns": None,
            "n_missing_total": None,
            "missing_percentage": None,
            "error": str(e)
        })

inventory_df = pd.DataFrame(inventory)
inventory_df.head()
```

```python
inventory_df.to_csv("../reports/tables/data_inventory.csv", index=False)
```

```python
sample_2023 = [f for f in excel_files if "2023.xlsx" in str(f)][0]
sample_2024 = [f for f in excel_files if "2024.xlsx" in str(f)][0]

df_2023 = pd.read_excel(sample_2023)
df_2024 = pd.read_excel(sample_2024)

display(df_2023.head())
```

## Variables del dataset
- **Año**: Año en el que se registró la observación.
- **Mes**: Mes del año en el que se tomó la medida.
- **Día**: Día del mes correspondiente al registro.
- **Hora**: Hora del día en la que se realizó la medición.
- **Minuto**: Minuto exacto en el que se realizó la medición.
- **Fecha**: Marca temporal completa correspondiente a la observación.
- **GHI**: Irradiancia solar global medida sobre una superficie horizontal.
- **DNI**: Irradiancia solar directa medida sobre una superficie perpendicular a los rayos solares.
- **DHI**: Irradiancia solar difusa recibida sobre una superficie horizontal.
- **Temperatura [°C]**: Temperatura ambiente registrada, expresada en grados Celsius.
- **Velocidad del viento [m/s]**: Velocidad del viento registrada, expresada en metros por segundo.
- **Dirección del viento [°]**: Dirección de procedencia del viento, expresada en grados.
- **Humedad relativa [%]**: Humedad relativa del aire en el momento de la medición, expresada como porcentaje.
- **Código GHI**: Variable objetivo con valores entre 0 y 1 que indica la calidad o validez de la medición de GHI.
- **Código DNI**: Variable objetivo con valores entre 0 y 2 que indica la calidad o validez de la medición de DNI.
- **Código DHI**: Variable objetivo con valores entre 0 y 2 que indica la calidad o validez de la medición de DHI.
